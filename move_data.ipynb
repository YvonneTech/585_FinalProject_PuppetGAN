```python
weizmann_path = './weizmann/'
save_path = './PuppetGAN/data/action/real_/real/'

count = 0

for fn in os.listdir(weizmann_path):
    for fn2 in os.listdir(weizmann_path+fn):
        if not fn2.endswith('.avi') and not fn2.startswith('.ipynb_checkpoints'):
            im_list = []
            path = weizmann_path + fn + '/' + fn2

            for fn3 in os.listdir(path):
                if fn3.endswith('.png'):
                    if len(str(count)) < 4:
                        str_count = '0' * (4 - len(str(count))) + str(count)
                    else:
                        str_count = str(count)

                    copy_from = path + '/' + fn3
                    copy_to = save_path + str_count + '.png'
                    # print(copy_from, copy_to)
                    shutil.copy(copy_from, copy_to)

                    count += 1
```
