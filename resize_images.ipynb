```python
import cv2 as cv
import os
```

```python
# 把synth_/synth的3连图 resize 256*768 -> 128*384
for img in os.listdir('./'):
    if img.endswith('.png'):
        orig_img = cv.imread(img)
        
        dim = (128, 384)

        resized_img = cv.resize(orig_img, dim)

        save_path = './' + img
        print(save_path)
        cv.imwrite(save_path, resized_img)
```

```python
src_path = './bodies/real_/real/'
dis_path = './bodies_32/real_/real/'
count = 0

for img in os.listdir(src_path):
    if img.endswith('.png'):
        orig_img = cv.imread(src_path+img)
        
        dim = (32, 32)

        resized_img = cv.resize(orig_img, dim)

        save_path = dis_path + img
        print(save_path)
        cv.imwrite(save_path, resized_img)

        count += 1

print(count)
```
