```python
borderType = cv.BORDER_CONSTANT
value = [255,255,255]

for img in os.listdir('./'):
    src = cv.imread("./" + img)
    padded_img = cv.copyMakeBorder(src, 0, 0, 0, 256, borderType, None, value)
    cv.imwrite("./"+img, padded_img)
    print("./"+img)
```
